```python
import numpy as np
import scipy
import networkx as nx
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter
import seaborn as sns
from scipy.optimize import curve_fit

from markov_epidemic import *
```

The goal of this notebook is to highlight the relationship between epidemic propagation and the topology of the underlying network of individuals; in particular, the denser the network, the faster and more critical the outbreak.

A lockdown policy corresponds to reducing the connectivity of the network, in particular by lowering its adjacency spectral radius.

## SEIR Epidemic

Consider a network $\mathcal{G}$ of $N$ individuals subject to an epidemic disease.
Let $A$ be the adjacency matrix of the underlying graph and $X_i(t)$ the health state of individual $i$ at time $t$.

We assume the following random Markov dynamic:

$$
\begin{cases} 
      \text{healthy} \rightarrow \text{exposed} & \text{at rate $\beta\sum_j A_{ij} \mathbb{1}_{X_j(t) = \text{infected}}$,}  \\
      \text{exposed} \rightarrow \text{infected} & \text{at rate $\gamma$.} \\
      \text{infected} \rightarrow \text{recovered} & \text{at rate $\delta$.} \\
   \end{cases}
$$

In other words, susceptible individuals are contaminated at a rate proportional to the number of infected neighbors in the network, then they undergo an incubation period at rate $\gamma$ during which they carry the disease but are not showing symptoms, and then recover at constant rate $\delta$. Once an individual has recovered it can never be infected again. Also we assume only one individual can change state at a given time $t$ (the events of multiple simultaneous transitions have probability zero anyway).

This is the **Susceptible-Exposed-Infected-Recovered (SEIR)** model.

## Sparse vs dense networks

```python
N = 500
G1 = nx.random_regular_graph(10, N)
G2 = nx.random_regular_graph(100, N)

# save layout
pos = nx.spring_layout(G1)

fig, axes = plt.subplots(figsize=(13, 5), nrows=1, ncols=2)

ax = axes[0]
nx.draw_networkx(G1, pos=pos, ax=ax, with_labels=False, node_size=1)
ax.set_title('Sparser graph')
ax.axis('off')

ax = axes[1]
nx.draw_networkx(G2, pos=pos, ax=ax, with_labels=False, node_size=1)
ax.set_title('Denser graph')
ax.axis('off')

plt.show()
```

### Epidemic parameters

```python
exposition_rate = 2.0
infection_rate = 0.5
recovery_rate = 1.5
epidemic1 = MarkovSEIR(exposition_rate, infection_rate, recovery_rate, G1)
epidemic2 = MarkovSEIR(exposition_rate, infection_rate, recovery_rate, G2)
```

```python
with sns.axes_style("darkgrid"):
    fig, axes = plt.subplots(figsize=(15, 13), nrows=2, ncols=1)

    ax = axes[0]
    ax.bar(range(1, N+1), epidemic1.spectrum, label='sparse', color='b', alpha=0.5)
    ax.bar(range(1, N+1), epidemic2.spectrum, label='dense', color='r', alpha=0.5)
    ax.set_title('Adjacency Spectrum')
    ax.legend(loc='lower left')
    
    ax = axes[1]
    isr1 = 1/epidemic1.spectral_radius
    edr = epidemic1.effective_diffusion_rate
    isr2 = 1/epidemic2.spectral_radius
    
    ax.axhline(y=isr1, color='b', label='Inverse spectral radius (sparse)')
    ax.axhline(y=isr2, color='r', label='Inverse spectral radius (dense)')
    ax.axhline(y=edr, color='g', label='Effective diffusion rate', linestyle='dashed')

    ax.set_xlim(0, 1)
    ax.set_ylim(np.min((edr, isr1, isr2))-0.2, np.max((edr, isr1, isr2))+0.1)
    ax.set_title('Epidemic threshold')
    ax.legend(loc='lower left')

    plt.show()
```

### Epidemic simulation

```python
T = 50.0
initial_infected = 5

epidemic1.simulate(T, epidemic1.random_seed_nodes(initial_infected))
epidemic2.simulate(T, epidemic2.random_seed_nodes(initial_infected))
```

### Smooth curve fitting

```python
def normal(x, mu, sigma, scale):
    return scale * np.exp(-(x-mu)**2/(2*sigma**2))

def lognormal(x, mu, sigma, scale):
     return scale * np.exp(-sigma * np.log((x + 1e-10) / mu)**2)
```

```python
try:
    # popt1_lognormal, _ = curve_fit(lognormal, epidemic1.transition_times, epidemic1.number_of_infected/epidemic1.N, p0=(0.5, 1.0, 1.0))
    popt1_normal, _ = curve_fit(normal, epidemic1.transition_times, epidemic1.number_of_infected/epidemic1.N, p0=(0.5, 1.0, 1.0))
    popt2_lognormal, _ = curve_fit(lognormal, epidemic2.transition_times, epidemic2.number_of_infected/epidemic2.N, p0=(0.5, 1.0, 1.0))

    # popt1_lognormal_exposed, _ = curve_fit(lognormal, epidemic1.transition_times, epidemic1.number_of_exposed/epidemic1.N, p0=(0.5, 1.0, 1.0))
    popt1_normal_exposed, _ = curve_fit(normal, epidemic1.transition_times, epidemic1.number_of_exposed/epidemic1.N, p0=(0.5, 1.0, 1.0))
    popt2_lognormal_exposed, _ = curve_fit(lognormal, epidemic2.transition_times, epidemic2.number_of_exposed/epidemic2.N, p0=(0.5, 1.0, 1.0))

    # lognormal1 = lambda x: lognormal(x, popt1_lognormal[0], popt1_lognormal[1], popt1_lognormal[2])
    normal1 = lambda x: normal(x, popt1_normal[0], popt1_normal[1], popt1_normal[2])
    lognormal2 = lambda x: lognormal(x, popt2_lognormal[0], popt2_lognormal[1], popt2_lognormal[2])

    # lognormal1_exposed = lambda x: lognormal(x, popt1_lognormal_exposed[0], popt1_lognormal_exposed[1], popt1_lognormal_exposed[2])
    normal1_exposed = lambda x: normal(x, popt1_normal_exposed[0], popt1_normal_exposed[1], popt1_normal_exposed[2])
    lognormal2_exposed = lambda x: lognormal(x, popt2_lognormal_exposed[0], popt2_lognormal_exposed[1], popt2_lognormal_exposed[2])
    
    fit_success = True
except:
    fit_success = False
    print('Regression failed')
```


### Plot comparison

```python
with sns.axes_style("darkgrid"):
    fig, axes = plt.subplots(figsize=(11, 15), nrows=3, ncols=1)

    ax = axes[0]
    ax.plot(epidemic1.transition_times, epidemic1.number_of_infected/epidemic1.N, color='b', label='sparse')
    ax.plot(epidemic2.transition_times, epidemic2.number_of_infected/epidemic2.N, color='r', label='dense')
    
    if fit_success:
        ax.plot(epidemic1.transition_times, 
                normal1(epidemic1.transition_times), 
                color='b', 
                linestyle='dashed', 
                label='normal fit',
                alpha=0.5,
               )
        ax.plot(epidemic2.transition_times, 
                lognormal2(epidemic2.transition_times), 
                color='r', 
                linestyle='dashed',
                label='lognormal fit',
                alpha=0.5,
               )

    ax.set_title('Total number of infected individuals')
    ax.set_xlabel('t')
    ax.yaxis.set_major_formatter(FuncFormatter(lambda y, _: '{:.0%}'.format(y)))
    ax.legend(loc='upper right')
    
    ax = axes[1]
    ax.plot(epidemic1.transition_times, epidemic1.number_of_exposed/epidemic1.N, color='b', label='sparse')
    ax.plot(epidemic2.transition_times, epidemic2.number_of_exposed/epidemic2.N, color='r', label='dense')
    
    if fit_success:
        ax.plot(epidemic1.transition_times, 
                normal1_exposed(epidemic1.transition_times), 
                color='b', 
                linestyle='dashed', 
                label='normal fit',
                alpha=0.5,
               )
        ax.plot(epidemic2.transition_times, 
                lognormal2_exposed(epidemic2.transition_times), 
                color='r', 
                linestyle='dashed',
                label='lognormal fit',
                alpha=0.5,
               )
    
    ax.set_title('Total number of exposed individuals')
    ax.set_xlabel('t')
    ax.yaxis.set_major_formatter(FuncFormatter(lambda y, _: '{:.0%}'.format(y)))
    ax.legend(loc='upper right')
    
    ax = axes[2]
    ax.plot(epidemic1.transition_times, epidemic1.number_of_recovered/epidemic1.N, color='b', label='sparse')
    ax.plot(epidemic2.transition_times, epidemic2.number_of_recovered/epidemic2.N, color='r', label='dense')
    
    ax.set_title('Total number of recovered individuals')
    ax.set_xlabel('t')
    ax.yaxis.set_major_formatter(FuncFormatter(lambda y, _: '{:.0%}'.format(y)))
    ax.legend(loc='upper right')
    
    plt.tight_layout()
    plt.show()
```

### Estimation of periodic behavior in the number of infected

```python
period_info1 = period_estimator(epidemic1.transition_times, 
                                epidemic1.number_of_infected,
                                interp_kind='previous',
                                savgol_window = 111,
                                savgol_polyorder = 3,
                                width_around_peak = 0.2,
                                distance_between_peaks=100,
                               )

xcorr1 = period_info1['xcorr']
xcorr_tt1 = period_info1['xcorr_tt']
xcorr_smooth1 = period_info1['xcorr_smooth']
peaks1 = period_info1['peaks']

period_info2 = period_estimator(epidemic2.transition_times, 
                                epidemic2.number_of_infected,
                                interp_kind='previous',
                                savgol_window = 111,
                                savgol_polyorder = 3,
                                width_around_peak = 0.2,
                                distance_between_peaks=100,
                               )

xcorr2 = period_info2['xcorr']
xcorr_tt2 = period_info2['xcorr_tt']
xcorr_smooth2 = period_info2['xcorr_smooth']
peaks2 = period_info2['peaks']
```

```python
with sns.axes_style("darkgrid"):
    fig, axes = plt.subplots(figsize=(11, 10), nrows=2, ncols=1)

    ax = axes[0]
    ax.plot(xcorr_tt1, xcorr1, label='Autocorrelogram', color='b', alpha=0.2)
    ax.plot(xcorr_tt1, xcorr_smooth1, label='Smoothed autocorrelogram', color='g')
    
    for peak1 in peaks1:
        ax.axvline(peak1, ymin=0.1, ymax=0.8, color='r', linestyle='dashed', alpha=0.5)

    ax.legend(loc='upper right')
    ax.yaxis.set_major_formatter(FuncFormatter(lambda y, _: '{:.0%}'.format(y)))
    ax.set_title('Period={:.2f}'.format(period_info1['period']))
    
    ax = axes[1]
    ax.plot(xcorr_tt2, xcorr2, label='Autocorrelogram', color='b', alpha=0.2)
    ax.plot(xcorr_tt2, xcorr_smooth2, label='Smoothed autocorrelogram', color='g')
    
    for peak2 in peaks2:
        ax.axvline(peak2, ymin=0.1, ymax=0.8, color='r', linestyle='dashed', alpha=0.5)

    ax.legend(loc='upper right')
    ax.yaxis.set_major_formatter(FuncFormatter(lambda y, _: '{:.0%}'.format(y)))
    ax.set_title('Period={:.2f}'.format(period_info2['period']))
    
plt.tight_layout()
plt.show()
```

### Conclusion

On an epidemic with fast contagionion but long incubation :
* the epidemic lifetime is rather long because the incubation introduces a significant delay in the system;
* the feedback loop introduced by the incubation period makes the number of infected indivuals more oscillatory than in the SIR model;
* on a sparse network, the number of simultaneously infected individuals remains on low; on a denser network, one can still observe epidemic bursts.


## Comparison with deterministic baseline

The corresponding deterministic system is :

* $\frac{dS}{dt} = -\bar{d}\beta\frac{IS}{N}$,
* $\frac{dE}{dt} = \bar{d}\beta\frac{IS}{N}-\gamma E$,
* $\frac{dI}{dt} = \gamma E-\delta I$,
* $\frac{dR}{dt} = \delta I$,

where $\bar{d}$ is the average degree of the network.

This can be seen as the expected evolution of the Markovian dynamic. However it lacks its flexibility in terms of network modelling : as it relies on an average degree assumption, it is only really accurate for regular graphs.

```python
exposition_rate = 2.0
infection_rate = 0.1
recovery_rate = 1.5
k = 10
T = 10.0
initial_infected = 50

G = nx.random_regular_graph(k, N)
epidemic = MarkovSEIR(exposition_rate, infection_rate, recovery_rate, G)

# Stochastic evolution
epidemic.simulate(T, epidemic.random_seed_nodes(initial_infected))

# Deterministic evolution
t, y = epidemic.deterministic_baseline(T, initial_infected, k)
```

```python
with sns.axes_style("darkgrid"):
    fig, axes = plt.subplots(figsize=(13, 7), nrows=2, ncols=1)

    ax = axes[0]
    ax.plot(epidemic.transition_times, epidemic.number_of_infected/epidemic.N, color='b', label='stochastic')
    ax.plot(t, y[2]/epidemic.N, label='deterministic', color='r')
    ax.set_xlabel('t')
    ax.yaxis.set_major_formatter(FuncFormatter(lambda y, _: '{:.0%}'.format(y)))
    ax.set_title('% of infected, deterministic vs stochastic model on a {}-regular graph'.format(k))
    ax.legend(loc='lower left')
    
    ax = axes[1]
    ax.plot(epidemic.transition_times, epidemic.number_of_exposed/epidemic.N, color='b', label='stochastic')
    ax.plot(t, y[1]/epidemic.N, label='deterministic', color='r')
    ax.set_xlabel('t')
    ax.yaxis.set_major_formatter(FuncFormatter(lambda y, _: '{:.0%}'.format(y)))
    ax.set_title('% of exposed, deterministic vs stochastic model on a {}-regular graph'.format(k))
    ax.legend(loc='lower left')
    
    plt.tight_layout()
    plt.show()
```
